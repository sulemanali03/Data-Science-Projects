## Line Graph

We can create a line graph to compare each country's GDP by year. Lets separate our data into different countries, and then plot them.

```python
chile = df[df['Country'].str.contains('Chile')]
china = df[df['Country'].str.contains('China')]
germany = df[df['Country'].str.contains('Germany')]
mexico = df[df['Country'].str.contains('Mexico')]
us = df[df['Country'].str.contains('United States of America')]
zimbabwe = df[df['Country'].str.contains('Zimbabwe')]
```

```python
plt.plot(chile['Year'], chile['GDP'])
plt.plot(china['Year'], china['GDP'])
plt.plot(germany['Year'], germany['GDP'])
plt.plot(mexico['Year'], mexico['GDP'])
plt.plot(us['Year'], us['GDP'])
plt.plot(zimbabwe['Year'], zimbabwe['GDP'])
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'US', 'Zimbabwe'])
plt.title('Country GDP by Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```

Now we can compare each country and their GDP against each other. Here are some observations we can make:

US has significantly higher GDP compared to the others, while Zimbabwe has the lowest. It looks like most, if not all, are trending positively. China overtook the second place spot around 2007, trading places with Germany. All countries seem to have taken some sort of dip in GDP around 2009. 

It looks like Chile, Mexico, and Zimbabwe are pretty stagnant from this graph - but is that true? It may just look this way because we are comparing polarizing GDP numbers. We can take a closer look by graphing each country and their GDP on their own.

```python
plt.figure(figsize = (20, 10))
plt.subplot(3, 6, 1)
plt.plot(chile['Year'], chile['GDP'], color = 'blue')
plt.title('Chile GDP by Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.subplot(3, 6, 2)
plt.plot(china['Year'], china['GDP'], color = 'orange')
plt.title('China GDP by Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.subplot(3, 6, 3)
plt.plot(germany['Year'], germany['GDP'], color = 'green')
plt.title('Germany GDP by Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.subplot(3, 6, 4)
plt.plot(mexico['Year'], mexico['GDP'], color = 'red')
plt.title('Mexico GDP by Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.subplot(3, 6, 5)
plt.plot(us['Year'], us['GDP'], color = 'purple')
plt.title('US GDP by Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.subplot(3, 6, 6)
plt.plot(zimbabwe['Year'], zimbabwe['GDP'], color = 'brown')
plt.title('Zimbabwe GDP by Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.subplots_adjust(wspace = 0.4)
plt.show()
plt.clf()
```

Here we can see each country's GDP by year. Just like we predicted, Chile, Mexico, and Zimbabwe's GDP wasn't stagnant over the years - it only looked like that because we were comparing the numbers to high GDP countries like the US and China. Now we can see that Chile's GDP has had a pretty steady growth. Meixco had a great growth too, but a massive dip around 2009 compared to the other's dips. Zimbabwe had a negative trend until 2009's dip. After, Zimbabwe's GDP has skyrocketed. China, US, and Zimbabwe's GDP have continued to trend positively while Chile, Germany, and Mexico have started to trend negatively.


Graphing each country on their own gives us a closer and much more accurate look at the trend of GDP from years 2000 to 2015.

## Scatterplot

Now lets create a scatterplot for our countries to compare GDP and life expectancy.

```python
sns.scatterplot(x = 'Life expectancy at birth (years)', y = 'GDP', data = df, hue = df['Country'])
plt.title('GDP x Life expectancy')
plt.legend()
plt.ylabel('GDP in Trillions')
plt.show()
plt.clf()
```

Interesting! Here are some observations we can make by looking at our scatterplot:

Zimbabwe has a large range of years in life expactancy. We can see that Zimbabwe has the lowest GDP and also the lowest life expectacy. On the other hand, US has the highest GDP and a pretty high life expectancy. Taking a look at all of the countries, if we draw lines of best fit, all of them would trend positively and we can assume that there is a correlation between GDP and Life expectancy in years.
